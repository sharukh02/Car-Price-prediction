```python
driver = webdriver.Chrome("C:\\Users\\ASUS\\chromedriver.exe")
```

### Scrapping the data from cardekho

```python
driver.get('https://www.cardekho.com')
```

```python
cities=[]
for i in driver.find_elements_by_xpath('//ul[@class="subMenuDeep"]//li//a'):
    cities.append(i.get_attribute('href'))
```

```python
cities
```

```python
cities = cities[10:19]
```

```python
car=[]
for i in cities[:9]:
    driver.get(i)
    time.sleep(1)
    for i in range(250):
        driver.execute_script("window.scrollBy(0,700)","")
        time.sleep(2)
    for i in driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]//a'):
        car.append(i.get_attribute('href'))
```

```python
car = pd.DataFrame({"Dataset_url":car})
```

```python
# Building list to store data
Location = []
Engine =[]
Name = []
Name1 = []
Varient1 = []
Insurance=[]
Owner=[]
Milege=[]
Power=[]
Transmission=[]
Price = []
Fuel = []
Driven=[]
```

```python
for i in car.url:
    driver.get(i)

Fetching locations from url by using if else
    if "Pune" in i:
        Location.append("Pune")
    
    elif "Ahmedabad" in i:
        Location.append("Ahmedabad")
        
    elif "Mumbai" in i:
        Location.append("Mumbai")
    
    elif "Bangalore" in i:
        Location.append("Bangalore")
    
    elif "Chennai" in i:
        Location.append("Chennai")
    
    elif "Kolkata" in i:
        Location.append("Kolkata")    
        
    elif "Hyderabad" in i:
        Location.append("Hyderabad")
    
    elif 'Noida'or 'New-Delhi' or'Ghaziabad,Gurgaon' in i:
        Location.append("Delhi")
    
    elif "Jaipur" in i:
        Location.append("Jaipur")
    
    else:
        Location.append("Other")
    
    time.sleep(2)
    
        
           
    try:
        power = driver.find_element_by_xpath('//*[@id="topspec"]/div[1]/div/div/div[2]/div[1]/div/section/div/div/ul/li[3]/div[2]')
        Power.append(power.text)
            
    except NoSuchElementException:
        try:
            power = driver.find_element_by_xpath('//*[@id="specification-wdght"]/div/div/div/ul/li[3]/div/div[2]')
            Power.append(power.text)
        except NoSuchElementException:
            Power.append('NA')
            pass
    
    try:
        price = driver.find_element_by_xpath('//span[@class="amount"]')
        Price.append(price.text)
            
    except NoSuchElementException:
        
        try:
            price = driver.find_element_by_xpath('//div[@class="priceSection"]')
            Price.append(price.text)
        except NoSuchElementException:
            Price.append("NA")
            pass
            

    try:
        # Append in separate
        name = driver.find_element_by_xpath('//span[@class="cardSubHead"]')
        Name.append(name.text)
        # only above
    
    except NoSuchElementException:
        
        try:
            name1 = driver.find_element_by_xpath('//div[@class="gsc_col-xs-12"]//h1')
            Name.append(name1.text)
        except NoSuchElementException:
            Name.append("NA")
            pass
            
        
    
    try:
        drive = driver.find_element_by_xpath('//*[@id="widget-Overview"]/div/div/div/div[1]/ul/li[3]/div/div')
        Driven.append(drive.text)
    
    except NoSuchElementException:
        try:
            drive = driver.find_element_by_xpath('//*[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[4]/div/div[2]')
            Driven.append(drive.text)
        except NoSuchElementException:
            Driven.append("NA")
            pass
    
    
    try:
        
        fuel = driver.find_element_by_xpath('//*[@id="widget-Overview"]/div/div/div/div[1]/ul/li[5]/div/div')
        Fuel.append(fuel.text)
    
    except NoSuchElementException:
               
        try:
            fuel = driver.find_element_by_xpath('//*[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[3]/div/div[2]')
            Fuel.append(fuel.text)
        except NoSuchElementException:
            Fuel.append("NA")
            pass
        
    try:
        owner = driver.find_element_by_xpath('//*[@id="widget-Overview"]/div/div/div/div[1]/ul/li[4]/div/div')
        Owner.append(owner.text)
    
    except NoSuchElementException:
        
        try:
            owner = driver.find_element_by_xpath('//*[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[6]/div/div[2]')
            Owner.append(owner.text)
        except NoSuchElementException:
            Owner.append("NA")
            pass
        
    try:
        milege = driver.find_element_by_xpath('//*[@id="topspec"]/div[1]/div/div/div[2]/div[1]/div/section/div/div/ul/li[1]/div[2]')
        Milege.append(milege.text)
    
    except NoSuchElementException:

        try:
            milege = driver.find_element_by_xpath('//*[@id="specification-wdght"]/div/div/div/ul/li[1]/div/div[2]')
            Milege.append(milege.text)
        except NoSuchElementException:
            Milege.append("NA")
            pass
            
    
    try:    
        ins = driver.find_element_by_xpath('//*[@id="widget-Overview"]/div/div/div/div[1]/ul/li[7]/div/div')
        Insurance.append(ins.text)
        
    except NoSuchElementException:
        try:
            ins = driver.find_element_by_xpath('//*[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[9]/div/div[2]')
            Insurance.append(ins.text)
        except NoSuchElementException:
            Insurance.append("NA")
            pass
    
    try:
    
        engine = driver.find_element_by_xpath('//*[@id="topspec"]/div[1]/div/div/div[2]/div[1]/div/section/div/div/ul/li[2]/div[2]')
        Engine.append(engine.text)
        
    except NoSuchElementException:

        try:
            engine = driver.find_element_by_xpath('//*[@id="specification-wdght"]/div/div/div/ul/li[2]/div/div[2]')
            Engine.append(engine.text)
        except NoSuchElementException:
            Engine.append("NA")
            pass
        
    try:
        trans = driver.find_element_by_xpath('//*[@id="widget-Overview"]/div/div/div/div[1]/ul/li[6]/div/div')
        Transmission.append(trans.text)

    except NoSuchElementException:
        
        try:
            trans = driver.find_element_by_xpath('//*[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[8]/div/div[2]')
            Transmission.append(trans.text)
        except NoSuchElementException:
            Transmission.append("NA")
            continue       
```

```python
df = pd.DataFrame({"Engine":Engine,"Brand Name":Name,"Location":Location,"Driven":Driven,"Insuarnce":Insurance,"Owner":Owner,
                           "Transmission":Transmission,"Milege":Milege,"Fuel Type":Fuel,"Max Power(In bhp)":Power,"Price":Price})
df.to_csv("Car_dataset1.csv")
```
